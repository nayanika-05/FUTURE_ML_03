```python
import re

def clean_text(text):
    text = str(text)
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    text = text.lower()
    text = text.split()
    text = " ".join(text)
    return text
```

```python
df['cleaned_text'] = df['Resume_str'].apply(clean_text)
```

```python
df[['cleaned_text']].head()
```

```python
df.columns
```

```python
skills = [
    "python", "java", "c++", "sql", "excel",
    "machine learning", "deep learning", "nlp",
    "tensorflow", "pytorch", "pandas", "numpy",
    "data analysis", "statistics", "communication"
]
```

```python
def extract_skills(text):
    text = str(text).lower()
    found_skills = []

    for skill in skills:
        if skill in text:
            found_skills.append(skill)

    return found_skills
```

```python
df['skills'] = df['cleaned_text'].apply(extract_skills)
```

```python
df[['cleaned_text', 'skills']].head()
```

```python
from collections import Counter

all_skills = sum(df['skills'], [])
Counter(all_skills).most_common(10)
```

```python
job_description = """
We are looking for a Python developer with experience in machine learning,
data analysis, SQL, and NLP. Candidate should know pandas, numpy, and
basic deep learning concepts.
"""
```

```python
job_clean = clean_text(job_description)
job_clean
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documents = df['cleaned_text'].tolist()
documents.append(job_clean)
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
job_vector = tfidf_matrix[-1]
resume_vectors = tfidf_matrix[:-1]
```

```python
scores = cosine_similarity(resume_vectors, job_vector)
df['score'] = scores
```

```python
df = df.sort_values(by='score', ascending=False)
df[['cleaned_text', 'skills', 'score']].head(10)
```

```python
df[['skills', 'score']].head(10)
```

```python
job_skills = extract_skills(job_clean)
job_skills
```

```python
def skill_gap(candidate_skills, job_skills):
    candidate_skills = set(candidate_skills)
    job_skills = set(job_skills)

    missing = job_skills - candidate_skills
    matched = job_skills & candidate_skills

    return list(matched), list(missing)
```

```python
df['matched_skills'] = df['skills'].apply(lambda x: skill_gap(x, job_skills)[0])
df['missing_skills'] = df['skills'].apply(lambda x: skill_gap(x, job_skills)[1])
```

```python
df[['skills', 'matched_skills', 'missing_skills', 'score']].head(10)
```

```python
df.to_csv("output/ranked_candidates.csv", index=False)
```

```python
import streamlit as st
import pandas as pd

st.set_page_config(page_title="AI Resume Screening", layout="wide")

st.title("📄 AI Resume Screening System")

df = pd.read_csv("output/ranked_candidates.csv")

df['score_percent'] = (df['score'] * 100).round(1)

def get_status(score):
    if score >= 70:
        return "🟢 Highly Recommended"
    elif score >= 40:
        return "🟡 Moderate Match"
    else:
        return "🔴 Low Match"

df['status'] = df['score_percent'].apply(get_status)

st.subheader("Top Ranked Candidates")

st.dataframe(
    df[['skills', 'score_percent', 'status', 'missing_skills']],
    use_container_width=True
)
```

```python
st.subheader("Match Score Visualization")

for i, row in df.iterrows():
    st.write(row['skills'])
    st.progress(int(row['score_percent']))
    st.write(row['status'])
```
